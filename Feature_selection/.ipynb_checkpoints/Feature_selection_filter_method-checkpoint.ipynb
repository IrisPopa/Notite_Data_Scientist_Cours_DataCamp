# Basic Filter Methods

* Constant Features that show single values in all the observations in the dataset. These features provide no information that allows ML models to predict the target

```python
X = data.drop(columns = ['CreditCard'], axis = 1)
y= data['CreditCard']
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=1)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
numerical_x_train = [column for column in X_train.columns if len(X_train[column].unique()) <= 2]
numerical_x_train
```

```python
# import and create the VarianceThreshold object.
from sklearn.feature_selection import VarianceThreshold
import numpy as np
vs_constant = VarianceThreshold(threshold=0)

# select the numerical columns only.
# numerical_x_train = X_train[X_train.select_dtypes([np.number]).columns]
numerical_x_train = X_train[[column for column in X_train.columns if len(X_train[column].unique()) > 3]]

# fit the object to our data.
vs_constant.fit(numerical_x_train)

# get the constant colum names.
constant_columns = [column for column in numerical_x_train.columns
                    if column not in numerical_x_train.columns[vs_constant.get_support()]]

# detect constant categorical variables.
# constant_cat_columns = [column for column in X_train.columns 
#                         if (X_train[column].dtype == "O" and len(X_train[column].unique())  == 1)]
constant_cat_columns = [column for column in X_train.columns if len(X_train[column].unique())  == 1]

# conctenating the two lists.
all_constant_columns = constant_cat_columns + constant_columns

# drop the constant columns
# X_train.drop(labels=all_constant_columns, axis=1, inplace=True)
# X_test.drop(labels=all_constant_columns, axis=1, inplace=True)
```

```python
all_constant_columns
```

* Quasi-Constant Features in which a value occupies the majority of the records

```python
# make a threshold for quasi constant.
threshold = 0.98

# create empty list
quasi_constant_feature = []

# loop over all the columns
for feature in X_train.columns:

    # calculate the ratio.
    predominant = (X_train[feature].value_counts() / np.float(len(X_train))).sort_values(ascending=False).values[0]
    
    # append the column name if it is bigger than the threshold
    if predominant >= threshold:
        quasi_constant_feature.append(feature)   
        
print(quasi_constant_feature)

# drop the quasi constant columns
# X_train.drop(labels=quasi_constant_feature, axis=1, inplace=True)
# X_test.drop(labels=quasi_constant_feature, axis=1, inplace=True)
```

* Duplicated Features, which is self-explanatory—the same feature.

```python
# transpose the feature matrice
train_features_T = X_train.T

# print the number of duplicated features
print(train_features_T.duplicated().sum())

# select the duplicated features columns names
duplicated_columns = train_features_T[train_features_T.duplicated()].index.values
print(duplicated_columns)

# drop those columns
# X_train.drop(labels=duplicated_columns, axis=1, inplace=True)
# X_test.drop(labels=duplicated_columns, axis=1, inplace=True)
```

# Correlation Filter Methods

```python
# Pearson correlation coefficient -it’s used to summarize the strength of the linear relationship between two data variables
# Spearman’s rank correlation coefficient - assesses monotonic relationships (whether linear or not) and is suitable for both continuous and discrete ordinal variables.
# Kendall’s rank correlation coefficient - measures the strength of the ordinal association between two variablesT and is best suited for discrete data
```

```python
# creating set to hold the correlated features
corr_features = set()

# create the correlation matrix (default to pearson)
corr_matrix = X_train.corr()

# optional: display a heatmap of the correlation matrix
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
plt.figure(figsize=(11,11))
sns.heatmap(corr_matrix, annot=True)
plt.show()

for i in range(len(corr_matrix .columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > 0.8:
            colname = corr_matrix.columns[i]
            corr_features.add(colname)
print(corr_features)

# X_train.drop(labels=corr_features, axis=1, inplace=True)
# X_test.drop(labels=corr_features, axis=1, inplace=True)
```

```python
# The corr() function takes a parameter called method, which allows us to specify the type of correlation. 
# By default it is the pearson correlation, but it can also be specified as kendall or spearman.
```

# Statistical & Ranking Filter Methods

* Mutual Information - measures the amount of information obtained about one variable through observing the other variable. In machine learning, mutual information measures how much information the presence/absence of a feature contributes to making the correct prediction on Y.

```python
# import the required functions and object.
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

# select the number of features you want to retain.
select_k = 5

# get only the numerical features.
# numerical_x_train = X_train[X_train.select_dtypes([np.number]).columns]
numerical_x_train = X_train[[column for column in X_train.columns if len(X_train[column].unique()) > 3]]

# create the SelectKBest with the mutual info strategy.
selection = SelectKBest(mutual_info_classif, k=select_k).fit(numerical_x_train, y_train)

# display the retained features.
features = numerical_x_train.select_dtypes([np.number]).columns[selection.get_support()]
print(features)
```

```python
numerical_x_train.columns
```

* Chi-squared Score - it’s suited for categorical variables and binary targets only, and the variables should be non-negative and typically boolean, frequencies, or counts.

```python
# import the required functions and object.
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

# change this to how much features you want to keep from the top ones.
select_k = 3

# select the numerical columns only.
categ_x_train = X_train[[column for column in X_train.columns if len(X_train[column].unique()) <= 3]]

# apply the chi2 score on the data and target (target should be binary).  
selection = SelectKBest(chi2, k=select_k).fit(categ_x_train, y_train)

# display the k selected features.
features = categ_x_train.columns[selection.get_support()]
print(features)
```

```python
categ_x_train.columns
```

* ANOVA Univariate Test - it’s well-suited for continuous variables and requires a binary target, but sklearn extends it to regression problems, also. The next example applies to classification tasks, but we can use ANOVA for regression tasks via the f_regression function provided in the sklearn library.

```python
# import the required functions and object.
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest

# select the number of features you want to retain.
select_k = 6

# create the SelectKBest with the mutual info strategy.
selection = SelectKBest(f_classif, k=select_k).fit(X_train, y_train)

# display the retained features.
features = X_train.columns[selection.get_support()]
print(features)
```

* Univariate ROC-AUC /RMSE - measure the dependence of two variables. It’s suitable for all variables. The measure here depends on the problem: we use RMSE for regression problems and ROC-AUC for classification problems

```python
# import the DecisionTree Algorithm and evaluation score.
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score

# list of the resulting scores.
roc_values = []

# loop over all features and calculate the score.
for feature in X_train.columns:
    clf = DecisionTreeClassifier()
    clf.fit(X_train[feature].to_frame(), y_train)
    y_scored = clf.predict_proba(X_test[feature].to_frame())
    roc_values.append(roc_auc_score(y_test, y_scored[:, 1]))

# create a Pandas Series for visualisation.
roc_values = pd.Series(roc_values)
roc_values.index = X_train.columns

# show the results.
print(roc_values.sort_values(ascending=False))
```

```python
# import the DecisionTree Algorithm and evaluation score.
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import mean_squared_error

# list of the resulting scores.
RMSE = []

# loop over all features and calculate the score.
for feature in X_train.columns:
    clf = DecisionTreeClassifier()
    clf.fit(X_train[feature].to_frame(), y_train)
    y_scored = clf.predict_proba(X_test[feature].to_frame())
    RMSE.append(mean_squared_error(y_test, y_scored[:, 1]))

# create a Pandas Series for visualisation.
RMSE = pd.Series(RMSE)
RMSE.index = X_train.columns

# show the results.
print(np.sqrt(roc_values).sort_values(ascending=True))
```
