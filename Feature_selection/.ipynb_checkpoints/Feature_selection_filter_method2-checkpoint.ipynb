## 2. Mutual Information

```python
# import the required functions and object.
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest

y_train = data["Industry"]
x_train = data.drop(columns=["Industry"])

# select the number of features you want to retain.
select_k = 8

# get only the numerical features.
numerical_x_train = x_train[x_train.select_dtypes([np.number]).columns]
print(numerical_x_train.shape)


# create the SelectKBest with the mutual info strategy.
selection = SelectKBest(mutual_info_classif, k=select_k).fit(numerical_x_train, y_train)

# display the retained features.
features = numerical_x_train.columns[selection.get_support()]
print(features)
```

## 3. Chi-squared Score

```python
# import the required functions and object.
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

# change this to how much features you want to keep from the top ones.
select_k = 8

# apply the chi2 score on the data and target (target should be binary).  
selection = SelectKBest(chi2, k=select_k).fit(numerical_x_train, y_train)

# display the k selected features.
features = numerical_x_train.columns[selection.get_support()]
print(features)
```

## 4. ANOVA Univariate Test

```python
# import the required functions and object.
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import SelectKBest

# select the number of features you want to retain.
select_k = 10

# create the SelectKBest with the mutual info strategy.
selection = SelectKBest(f_classif, k=select_k).fit(numerical_x_train, y_train)

# display the retained features.
features = numerical_x_train.columns[selection.get_support()]
print(features)
```

```python
# import the DecisionTree Algorithm and evaluation score.
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score

# list of the resulting scores.
roc_values = []

# loop over all features and calculate the score.
for feature in x_train.columns:
    clf = DecisionTreeClassifier()
    clf.fit(x_train[feature].to_frame(), y_train)
    y_scored = clf.predict_proba(X_test[feature].to_frame())
    roc_values.append(roc_auc_score(y_test, y_scored[:, 1], multi_class="ovr"))

# create a Pandas Series for visualisation.
roc_values = pd.Series(roc_values)
roc_values.index = X_train.columns

# show the results.
print(roc_values.sort_values(ascending=False))
```
