```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = tf.linspace(-2,2,201)
x = tf.cast(x, tf.float32)

def f(x):
  y = x ** 2 + 2 * x - 5
  return y

y = f(x) + tf.random.normal(shape=[201])

plt.plot(x.numpy(), y.numpy(), '.', label="Data")
plt.plot(x, f(x), label="Ground truth")
plt.legend()
```

```python
class Model(tf.keras.Model):
  def __init__(self, units, num_lyrs):
    super().__init__()
    self.dense1 = tf.keras.layers.Dense(
        units=units,
        activation=tf.nn.relu,
        kernel_initializer=tf.random.normal,
        bias_initializer=tf.random.normal
      )
    self.dense2 = tf.keras.layers.Dense(1)
  
  def call(self, x, training=True):
    x = x[:, tf.newaxis]
    x = self.dense1(x)
    x = self.dense2(x)
    return tf.squeeze(x, axis=1)

model = Model(10)
model.build(input_shape=[1])
model.summary()
variables = model.variables
optimizer = tf.optimizers.SGD(learning_rate=0.01)
for step in range(1000):
  with tf.GradientTape() as tape:
    prediction = model(x)
    error = (y-prediction)**2
    mean_error = tf.reduce_mean(error)
  if step % 100 == 0:
    print("Mean squared error: ", mean_error.numpy() )
  gradient = tape.gradient(mean_error, variables)
  optimizer.apply_gradients(zip(gradient, variables))
```

```python
plt.plot(x.numpy(), y.numpy(), ".", label="data")
plt.plot(x, f(x), label="Ground truth")
plt.plot(x, model(x), label="Trained predictions")
plt.title("After training")
plt.legend();
```

```python
class Model(tf.keras.Model):
  def __init__(self, units, num_layers=10):
    super().__init__()
    self.lyrs = []
    for i in range(num_layers):
      self.lyrs.append(tf.keras.layers.Dense(
                              units=units,
                              activation=tf.nn.relu,
                              kernel_initializer=tf.random.normal,
                              bias_initializer=tf.random.normal
                            ))

    self.dense2 = tf.keras.layers.Dense(1)
  
  def call(self, x, training=True):
    x = x[:, tf.newaxis]
    for l in self.lyrs:
      x = l(x)
    x = self.dense2(x)
    return tf.squeeze(x, axis=1)

model = Model(units=10,num_layers=2)
model.build(input_shape=[1])
variables = model.variables
optimizer = tf.optimizers.SGD(learning_rate=0.000001)
for step in range(1000):
  with tf.GradientTape() as tape:
    prediction = model(x)
    error = (y-prediction)**2
    mean_error = tf.reduce_mean(error)
  if step % 100 == 0:
    print("Mean squared error: ", mean_error.numpy() )
  gradient = tape.gradient(mean_error, variables)
  optimizer.apply_gradients(zip(gradient, variables))
```

```python
plt.plot(x.numpy(), y.numpy(), ".", label="data")
plt.plot(x, f(x), label="Ground truth")
plt.plot(x, model(x), label="Trained predictions")
plt.title("After training")
plt.legend();
```
